## Sub plot and legends

```python
days = [1,2,3,4,5]
original_price =[10,20,25,30,40]
predicted_price=[8,21,25,28,38]
```

```python
plt.plot(days,original_price,'b-o', label='Original Price')
plt.plot(days,predicted_price,'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('WeeksDays')
plt.ylabel('Price')
```

### Subplots

```python
# First Subplots-original price
plt.subplot(3,1,1)
plt.plot(days,original_price,'b-o', label='Original Price')
plt.title('Original Prices over time ')
plt.xlabel('WeeksDays')
plt.ylabel('Price')
plt.legend()

# First Subplots-original price
plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o', label='Predicted Price')
plt.title('Predicted Prices over time ')
plt.xlabel('WeeksDays')
plt.ylabel('Price')
plt.legend()



# Third Subplots-Difference Between original & Predicted price
price_difference = [o-p for o, p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.bar(days,price_difference, color='purple', label='Price diference')
plt.title('Difference Between original & Predicted price')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()

plt.tight_layout()
```

## Saving the figure

```python
plt.plot(days,original_price,'b-o', label='Original Price')
plt.plot(days,predicted_price,'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('WeeksDays')
plt.ylabel('Price')

plt.savefig('example_prediction.png',dpi=300)  # save the image
```
